```python
x, y = sym.symbols("x y")
eq1 = x**2+3*y-10
eq2 = 4*x-y**2+2
solutions = solve([eq1, eq2])
solutions
```

```python
for s in solutions:
    for k, v in s.items():
        print(k, v.evalf())

```
